```python
import pandas as pd
import numpy as np
```

```python
# Load Q4 impressions
impressions_df = pd.read_csv("../datasets/impressions_q4_extended.csv")
impressions_df.head()
```

```python
# Load schibsted report
schibsted_df = pd.read_excel("../datasets/report - 2023-02-01T152103.748 (1).xlsx")
schibsted_df
```

```python
import json

def extract_pid(placement_id):
    placement_id = json.loads(placement_id)
    bam_ad_slots = placement_id["bam_ad_slot"]
    # print(bam_ad_slots)
    final_bam_ad_slots = [bas for bas in bam_ad_slots if bas.strip() != '']
    ret = ''
    try:
        ret = placement_id['tag_id'][0]
    except:
        # print('.')
        ret = ','.join([bas for bas in final_bam_ad_slots])
    else:
        ret += ','.join([bas for bas in final_bam_ad_slots])
    return ret
    

def len_pids(placement_id):
    placement_id = json.loads(placement_id)
    bam_ad_slots = placement_id["bam_ad_slot"]
    final_bam_ad_slots = [bas for bas in bam_ad_slots if bas.strip() != '']
    return len(placement_id["tag_id"]) + len(final_bam_ad_slots)

def pid_type(pid):
    try:
        int(pid)
        return 'int'
    except:
        return 'str'

impressions_df['pid'] = impressions_df['placement_ids'].apply(extract_pid)
impressions_df['pid_len'] = impressions_df['placement_ids'].apply(len_pids)
impressions_df['pid_type'] = impressions_df['pid'].apply(pid_type)
```

```python
impressions_df = impressions_df[impressions_df["pid_type"] == 'int']
impressions_df['pid'] = impressions_df['pid'].apply(lambda pid: int(pid))
impressions_df
```

```python
pids = list(schibsted_df["Placement Id"].unique())
# Filter impressions to just include PID's from report
ab_df = impressions_df[impressions_df["pid"].isin(pids)]
ab_df
```

```python
# Filter impression set to only use expressen
# client = "expressen.se"
#client_df = impressions_df[impressions_df["hostname"] == client]
# Get bonnier impression data
clients = ["expressen.se", "dn.se", "svd.se"]
bonnier_df = impressions_df[impressions_df["hostname"].isin(clients)]
bonnier_df
```

```python
client_df = pd.concat([bonnier_df, ab_df])
client_df
```

```python
client_df = client_df[client_df["pid_type"] == "int"]
```

```python
client_df["hostname"].value_counts()
```

```python
grouped_df = client_df.groupby(["hostname", "pid"]).apply(
    lambda x: pd.Series(
        {
            "num_impressions": x["id"].count(),
            "is_iab_inview": x[x["is_iab_inview"] == True]["id"].count()
        }
    )
)
grouped_df = grouped_df.reset_index()
grouped_df
```

```python
grouped_df["tobii_viewability%"] = grouped_df.apply(lambda x: round(x["is_iab_inview"] / x["num_impressions"] * 100, 0), axis=1)
grouped_df
```

```python
grouped_df["hostname"].value_counts()
```

```python
average_df = grouped_df.groupby("hostname").apply(
    lambda x: pd.Series(
        {
            "average_tobii_viewability": np.average(x["tobii_viewability%"]),
            "num_of_placements": x["pid"].count(),
            "num_of_impressions": np.sum(x["num_impressions"])
        }
    )
)
average_df = average_df.reset_index()
average_df
```

```python
average_df.to_excel("../final_reports/ATEX2-965_average_stats.xlsx")
average_df.to_csv("../final_reports/ATEX2-965_average_stats.csv")

grouped_df.to_excel("../final_reports/ATEX2-965_tobii_viewability_comparison.xlsx")
grouped_df.to_csv("../final_reports/ATEX2-965_tobii_viewability_comparison.csv")
print("done")
```
