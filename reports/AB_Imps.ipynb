```python
import pandas as pd
import numpy as np
```

```python
impressions_df = pd.read_csv("/home/sk2159/Downloads/67ca8694-feba-4b35-bc94-13c6384de43a.csv")
```

```python
# run through predictions
import os
import json


def _get_pred_map():
    pred_map = dict()
    for i in range(3, 6):
        if i < 10:
            i = f'0{i}'
        for day in os.listdir(f'../predictions/2023/{i}'):
            for h in os.listdir(f'../predictions/2023/{i}/{day}'):
                if not h.endswith('.ndjson'):
                    continue
                with open(f'../predictions/2023/{i}/{day}/{h}')as f:
                    for line in f.readlines():
                        json_line = json.loads(line)
                        id_ = json_line['id']
                        pred_map[id_] = json_line['prediction']
    return pred_map
                            
pred_map = _get_pred_map()
```

```python
def _get_final_fixation(gaze_valid, is_fixated, id_, pred_map):
    if gaze_valid:
        return is_fixated
    if id_ in pred_map:
        return pred_map[id_]
    return False

impressions_df['final_fixation'] = impressions_df.apply(lambda row: _get_final_fixation(row['gaze_valid'], row['is_fixated'], row['id'], pred_map), axis=1)
impressions_df
```

```python
impressions_df["true_is_iab_inview"] = impressions_df.apply(
    lambda x: 
    True if (
        (x["ad_technical_format"] == "out-stream") & (x["exist_viewable_1_s_threshold_50"] == True) & (x["exist_viewable_2_s_threshold_50"] == False)
    ) | (x["is_iab_inview"] == True) 
    else False, 
    axis=1
)
```

```python
final_group = impressions_df.groupby(["hostname"]).apply(
     lambda x: pd.Series(
         {
             "impressions": x["id"].count(),
             "fixations": x.loc[x["final_fixation"], "id"].count(),
             "inviews": x.loc[x["true_is_iab_inview"], "id"].count(),
         }
     )
)
final_group = final_group.reset_index()
final_group["inview ratio"] = round(final_group["inviews"] / final_group["impressions"] * 100, 2)
final_group["fixation ratio"] = round(final_group["fixations"] / final_group["impressions"] * 100, 2)
final_group["fixation/inview ratio"] = round(final_group["fixations"] / final_group["inviews"] * 100, 2)
```
