```python
import pandas as pd
import numpy as np
```

```python
# load impressions for 2022
impressions_df = pd.read_csv('../datasets/impressions_10_11_12.csv')
impressions_df
```

```python
import json

def extract_pid(placement_id):
    placement_id = json.loads(placement_id)
    print(placement_id)
    bam_ad_slots = placement_id["bam_ad_slot"]
    final_bam_ad_slots = [bas for bas in bam_ad_slots if bas.strip() != '']
    try:
        ret = placement_id['tag_id'][0]
    except:
        print('empty pid')
        return
    ret += ','.join([bas for bas in final_bam_ad_slots])
    return ret
    

def len_pids(placement_id):
    placement_id = json.loads(placement_id)
    bam_ad_slots = placement_id["bam_ad_slot"]
    final_bam_ad_slots = [bas for bas in bam_ad_slots if bas.strip() != '']
    return len(placement_id["tag_id"]) + len(final_bam_ad_slots)

def pid_type(pid):
    try:
        int(pid)
        return 'int'
    except:
        return 'str'

impressions_df['pid'] = impressions_df['placement_ids'].apply(extract_pid)
impressions_df['pid_len'] = impressions_df['placement_ids'].apply(len_pids)
impressions_df['pid_type'] = impressions_df['pid'].apply(pid_type)
```

```python
impressions_df['pid_type'].value_counts()
```

```python
# filter only impressions with int placement IDs
impressions_df = impressions_df[impressions_df['pid_type'] == 'int']
impressions_df.head()
```

```python
# convert placement ids that can be converted to the int to be int, not string
# type(impressions_df.iloc[0]['pid'])     # check this, right now they're strings
impressions_df['pid'] = impressions_df['pid'].apply(lambda pid: int(pid))
```

```python
type(impressions_df.iloc[0]['pid'])
```

```python
# run through predictions
import os
import json


def _get_pred_map():
    pred_map = dict()
    
    for i in range(1, 13):
        if i < 10:
            i = f'0{i}'
        for day in os.listdir(f'../predictions/2022/{i}'):
            for h in os.listdir(f'../predictions/2022/{i}/{day}'):
                if not h.endswith('.ndjson'):
                    continue
                with open(f'../predictions/2022/{i}/{day}/{h}')as f:
                    for line in f.readlines():
                        json_line = json.loads(line)
                        id_ = json_line['id']
                        pred_map[id_] = json_line['prediction']
    return pred_map
                            
pred_map = _get_pred_map()
```

```python
def _get_final_fixation(gaze_valid, is_fixated, id_, pred_map):
    if gaze_valid:
        return is_fixated
    if id_ in pred_map:
        return pred_map[id_]
    return False

impressions_df['final_fixation'] = impressions_df.apply(lambda row: _get_final_fixation(row['gaze_valid'], row['is_fixated'], row['id'], pred_map), axis=1)
impressions_df
```

```python
# load tele2 dataset
df_client_tele2 = pd.read_csv('../datasets/bq-results-20221214-130759-1671023289214.csv')
df_client_tele2
```

```python
# filter only impressions
df_client_tele2 = df_client_tele2[df_client_tele2['is_imp'] == True]
df_client_tele2.shape
```

```python
df_client_tele2['testgroup'].value_counts()
```

```python
# filter specific campaign
campaign = 'control_group'
df_client_tele2 = df_client_tele2[df_client_tele2['testgroup'] == campaign]
print(df_client_tele2.shape)
```

```python
# filter only impressions with tele2 placement IDs
tele2_placements = list(df_client_tele2['placement_id'].unique())
tele2_impressions = impressions_df[impressions_df['pid'].isin(tele2_placements)]
tele2_impressions
```

```python
len(tele2_placements)
```

```python
# groupby and aggregate
grouped_df = tele2_impressions.groupby(['hostname', 'pid']).apply(
    lambda x: pd.Series(
        {
            'num_impressions' : x['id'].count(),
            'num_fixations': x.loc[x['final_fixation'],'id'].count()
        }
    )
)
```

```python
grouped_df['ratio'] = round(grouped_df['num_fixations'] / grouped_df['num_impressions'], 4)
grouped_df['ratio_percent'] = round(grouped_df['ratio'] * 100, 2)
grouped_df
```

```python
tele2_grouped_df[tele2_grouped_df['num_impressions'] != tele2_grouped_df['num_inview']]
```

```python
def get_sample_size(num_impressions):
    if num_impressions <= 5:
        return "low"
    elif num_impressions > 5 and num_impressions <= 10:
        return "medium"
    else:
        return "high"

# merge grouped_df with tele2_grouped_df to have number of impressions from tele2 input data
new_df = pd.merge(grouped_df, tele2_grouped_df,  how='left', left_on=['hostname','pid'], right_on = ['hostname','placement_id'])
new_df.rename(columns={"num_impressions_y": "tele2_impressions"}, inplace=True)
new_df['tele2_fixations'] = round(new_df['tele2_impressions'] * new_df['ratio'])
# recalculate ratio and ratio_percent, to be exact value from their data
new_df['ratio'] = round(new_df['tele2_fixations'] / new_df['tele2_impressions'], 4)
new_df['ratio_percent'] = round(new_df['ratio'] * 100, 2)
new_df['sample_size'] = new_df['num_impressions_x'].apply(lambda num_impressions: get_sample_size(num_impressions))
```

```python
result_df = new_df[['hostname', 'placement_id', 'tele2_impressions', 'num_inview', 'tele2_fixations', 'ratio', 'ratio_percent', 'sample_size']]
result_df.rename(columns={'tele2_impressions': 'num_impressions', 'tele2_fixations': 'num_fixations'}, inplace=True)
result_df
```
