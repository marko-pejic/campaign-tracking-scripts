```python
import pandas as pd
import numpy as np
```

```python
client_df = pd.read_excel("../datasets/Schibsted_Placement_Viewability.xlsx")
report = pd.read_excel("../final_reports/ATEX2-966.xlsx", index_col=0)
```

```python
part_df = client_df[client_df["Placement Id"].isin(list(report["Placement Id"]))]
part_df
```

```python
part_df["Placement Id"].value_counts()
```

```python
merged_df = pd.merge(left=report, right=client_df, how="outer", on="Placement Id", indicator=True)
final_df = merged_df.dropna()
```

```python
final_df = final_df[["hostname", "Placement Id", "Placement Name", "Size","Viewability Measurement Rate %", "Viewability Rate % Xandr", "Viewability rate % Tobii IAB or 1 second", "Viewability Rate %", "_merge"]]
final_df
```

```python
final_df["Viewability Rate %"] = round(final_df["Viewability Rate %"] * 100, 0)
final_df = final_df.astype({"Viewability Measurement Rate %": "int64", "Viewability Rate % Xandr": "int64", "Viewability rate % Tobii IAB or 1 second": "int64", "Viewability Rate %": "int64",})
final_df
```

```python
final_df = final_df[["hostname", "Placement Id", "Placement Name", "Size","Viewability Measurement Rate %", "Viewability Rate % Xandr", "Viewability rate % Tobii IAB or 1 second", "Viewability Rate %"]]
final_df.to_excel("../final_reports/viewability_report_schibsted_vs_tobii_vs_xandr_vs_acceleration.xlsx", index=False)
```
