```python
impressions_df.rename(columns={'id': 'impression_id'}, inplace=True)
impressions_df
```

```python
# run through predictions
import os
import json


def _get_pred_map():
    preds = []
    
    for i in range(1, 2):
        print(i)
        if i not in (1, 11, 12):
            continue
        if i < 10:
            i = f'0{i}'
        year = 2023 if i == '01' else 2022
        for day in os.listdir(f'../predictions/{year}/{i}'):
            print(f'{year}/{i}/{day}')
            for h in os.listdir(f'../predictions/{year}/{i}/{day}'):
                if not h.endswith('.ndjson'):
                    continue
                with open(f'../predictions/{year}/{i}/{day}/{h}')as f:
                    for line in f.readlines():
                        json_line = json.loads(line)
                        id_ = json_line['id']
                        preds.append({
                            "impression_id": id_,
                            'prediction': json_line['prediction'],
                            'prediction_proba': json_line['prediction_proba'],
                            'training_years': json_line['training_years'],
                            'training_months': json_line['training_months']
                        })
    return preds
                            
pred_map = _get_pred_map()
```

```python
pred_map
```

```python
predictions_df = pd.DataFrame(pred_map, columns=['impression_id', 'prediction', 'prediction_proba', 'training_years', 'training_months'])
predictions_df
```

```python
result_df = impressions_df.merge(predictions_df, on='impression_id', how='inner')
result_df
```
