```python
import pandas as pd
import numpy as np
```

```python
impressions = pd.read_csv("../datasets/impressions/2023/impressions_q1.csv")
impressions_q4 = pd.read_csv("../datasets/impressions/2022/impressions_q4_extended_v2.csv")
#impressions = impressions[impressions["gaze_valid"] == True]
```

```python
impressions = impressions[(impressions["is_fixated"] == True) & (impressions["gaze_valid"] == True) & (impressions["total_fixation_duration"] <= 30000)]
impressions_q4 = impressions_q4[(impressions_q4["is_fixated"] == True) & (impressions_q4["gaze_valid"] == True) & (impressions_q4["total_fixation_duration"] <= 30000)]
```

```python
grouped_impressions_2023 = impressions.groupby(["channel"]).apply(
    lambda x: pd.Series(
        {
            "num_impressions": x["id"].count(),
            "num_fixations": x.loc[x["is_fixated"], "id"].count(),
            "avg_fixation_duration": np.average(x["total_fixation_duration"])
        }
    )
)
grouped_impressions_2023["avg_fixation_duration"] = grouped_impressions_2023["avg_fixation_duration"].apply(lambda x: x / 1000)
grouped_impressions_2023
```

```python
grouped_impressions_2022 = impressions_q4.groupby(["channel"]).apply(
    lambda x: pd.Series(
        {
            "num_impressions": x["id"].count(),
            "num_fixations": x.loc[x["is_fixated"], "id"].count(),
            "avg_fixation_duration": np.average(x["total_fixation_duration"])
        }
    )
)
grouped_impressions_2022["avg_fixation_duration"] = grouped_impressions_2022["avg_fixation_duration"].apply(lambda x: x / 1000)
grouped_impressions_2022
```
