```python
import pandas as pd
import numpy as np
import json
import re
from pyathena import connect
from pyathena.pandas.util import as_pandas
from pyathena.pandas.cursor import PandasCursor
import multiprocessing
import concurrent
```

```python
country = "Norway"
c_code = "no"
```

```python
client_df = pd.read_excel("../datasets/domain-data-dk-no.xlsx", sheet_name=country)
client_placement_ids = list(client_df["Placement Id"].unique())
```

```python
def run_query(placement_id):
#for i, placement_id in enumerate(client_placement_ids):
    #if i % 10 == 0:
    #    print(i)
    cursor.execute(f'''SELECT 
        "impression_id",
        "is_fixated",
        "is_iab_inview",
        {placement_id} as pid,
        CONCAT(cast(width as VARCHAR), 'x', cast(height as VARCHAR)) AS size
    FROM "data_pipelines_storage_nomarket_prod"."extension_impression"
    WHERE regexp_like(json_format(cast(transform(html_metadata, x -> ROW(x."inline_scripts", x."non_standard_attributes")) as JSON)), '(\D)({placement_id})(\D)')
    ''')

    return as_pandas(cursor)
```

```python
num_cores = multiprocessing.cpu_count()
total_count = len(client_placement_ids)
cursor = connect(s3_staging_dir="s3://adhoc-athena-output-bucket-eu-north-1/",
                 region_name="eu-north-1", profile_name="atexprodadminsso", cursor_class=PandasCursor).cursor()
impressions_list = list()

# try to execute in parallel
with concurrent.futures.ThreadPoolExecutor(max_workers=num_cores) as executor:
    for i, result in enumerate(executor.map(run_query, client_placement_ids), 1):
        print(f"Processing {i} out of {total_count}")
        impressions_list.append(result)

impressions_df = pd.concat(impressions_list)
```

```python
grouped_df = impressions_df.groupby(["pid", "size"]).apply(
    lambda x: pd.Series(
        {
            "impressions": x["impression_id"].count(),
            "fixations": x[x["is_fixated"] == True]["impression_id"].count(),
            "inview": x[x["is_iab_inview"] == True]["impression_id"].count()
        }
    )
)
grouped_df = grouped_df.reset_index()
```

```python
grouped_df["fixation_ratio"] = round(grouped_df["fixations"] / grouped_df["impressions"] * 100, 2)
grouped_df["inview_ratio"] = round(grouped_df["inview"] / grouped_df["impressions"] * 100, 2)
grouped_df
```

```python
merged_df = pd.merge(client_df, grouped_df, left_on=["Placement Id", "Size"], right_on=["pid", "size"], how="outer")
merged_df.head()
```

```python
updated_client_df = merged_df[["Seller", "Seller Name", "Publisher Id", "Publisher Name", "Placement Group Id", "Site Name", "Placement Id", "Placement Name", "Media Type Id", "Size", "impressions", "fixations", "inview","fixation_ratio", "inview_ratio"]]
```

```python
updated_client_df["fixation_ratio"] = updated_client_df["fixation_ratio"].fillna(0)
updated_client_df["inview_ratio"] = updated_client_df["inview_ratio"].fillna(0)
updated_client_df["impressions"] = updated_client_df["impressions"].fillna(0)
updated_client_df["fixations"] = updated_client_df["fixations"].fillna(0)
updated_client_df["inview"] = updated_client_df["inview"].fillna(0)
updated_client_df = updated_client_df.dropna(subset=['Seller'])
```
