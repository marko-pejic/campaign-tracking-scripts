```python
import pandas as pd
import numpy as np
```

```python
user_df = pd.read_excel("/home/sk2159/Downloads/Daily panel activity March.xlsx")
user_df
```

```python
user_df_only_emails = user_df.dropna(axis=0, subset=["email"])
user_df_only_emails
```

```python
grouped_df = user_df_only_emails.groupby(["date_column", "email", "eye_sn"]).apply(
    lambda x: pd.Series(
        {
            "Norstat_UK": x["Norstat UK"].any(),
        }
    )
)
grouped_df = grouped_df.reset_index()
grouped_df
```

```python
final_df = grouped_df[(grouped_df["eye_sn"] != "undefined") & (grouped_df["eye_sn"] != "?")]
```

```python
final_df.to_excel("../final_reports/march_daily_user_activity.xlsx", index=False)
```
