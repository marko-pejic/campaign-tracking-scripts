```python
import pandas as pd
import numpy as np
```

```python
impressions_df = pd.read_csv("./YT_imps_2023_05.csv")
impressions_df
```

```python
impressions_df["ad_technical_format"].value_counts()
```

```python
impressions_display = impressions_df[impressions_df["ad_technical_format"] == "display"][:20]
impressions_video_instream = impressions_df[impressions_df["ad_technical_format"] == "in-stream"][:20]
impressions_video_outstream = impressions_df[impressions_df["ad_technical_format"] == "out-stream"][:20]
final_df = pd.concat([impressions_display, impressions_video_instream, impressions_video_outstream])
```

```python
final_df.to_excel("./youtube_impressions.xlsx", index=False)
```
