```python
import pandas as pd
import numpy as np
```

```python
impressions_df = pd.read_csv("./YT_imps_2023_05.csv")
```

```python
impressions_df["ad_technical_format"].value_counts()
```

```python
impressions_display_is_fixated = impressions_df[(impressions_df["ad_technical_format"] == "display") & (impressions_df["is_fixated"] == True)][:10]
impressions_display_not_fixated = impressions_df[(impressions_df["ad_technical_format"] == "display") & (impressions_df["is_fixated"] == False)][:10]
impressions_display = pd.concat([impressions_display_is_fixated, impressions_display_not_fixated])

impressions_video_instream_is_fixated = impressions_df[(impressions_df["ad_technical_format"] == "in-stream") & (impressions_df["is_fixated"] == True)][:10]
impressions_video_instream_is_not_fixated = impressions_df[(impressions_df["ad_technical_format"] == "in-stream") & (impressions_df["is_fixated"] == False)][:10]
impressions_video_instream = pd.concat([impressions_video_instream_is_fixated, impressions_video_instream_is_not_fixated])

impressions_video_outstream_is_fixated = impressions_df[(impressions_df["ad_technical_format"] == "out-stream") & (impressions_df["is_fixated"] == True)][:10]
impressions_video_outstream_is_not_fixated = impressions_df[(impressions_df["ad_technical_format"] == "out-stream") & (impressions_df["is_fixated"] == False)][:10]
impressions_video_outstream = pd.concat([impressions_video_outstream_is_fixated, impressions_video_outstream_is_not_fixated])

final_df = pd.concat([impressions_display, impressions_video_instream, impressions_video_outstream])
final_df
```

```python
final_df.to_excel("./youtube_impressions.xlsx", index=False)
```
