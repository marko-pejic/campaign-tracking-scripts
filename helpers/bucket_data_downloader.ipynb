```python
import boto3
import os

# Let's use Amazon S3
s3_resource = boto3.resource("s3")
s3_client = boto3.client('s3')
```

```python
bucket = 'prod-sov-consumer-events-atex-sov-eu-west-1'
year = '2023'
month = '01'
```

```python
def download_predictions(bucket, year, month):
    prefix = f'predictions/{year}/{month}'
    objects_response = s3_client.list_objects_v2(Bucket=bucket, Prefix=prefix)
    objects = objects_response['Contents']
    
    obj_name = None
    
    for obj in objects:
        obj_name = obj['Key']

        if obj_name.endswith('.ndjson'):
            # 9 is "00.ndjson"
            dir_path = obj_name[:-10]
            if not os.path.exists(dir_path):
                print(dir_path)
                os.makedirs(dir_path)
            
            print(obj_name)
            # first parameter: s3 source path, second parameter: local path
            s3_resource.Bucket(bucket).download_file(obj_name, obj_name)
```

```python
download_predictions(bucket, year, month)
```

```python
def download_impressions(bucket, year, month):
    prefix = f'impressions/{year}/{month}'
    objects_response = s3_client.list_objects_v2(Bucket=bucket, Prefix=prefix)
    objects = objects_response['Contents']
    
    obj_name = None
    
    for obj in objects:
        obj_name = obj['Key']

        if obj_name.endswith('.tar'):
            # 9 is "00.ndjson"
            dir_path = obj_name[:-7]
            if not os.path.exists(dir_path):
                print(dir_path)
                os.makedirs(dir_path)
            
            print(obj_name)
            # first parameter: s3 source path, second parameter: local path
            s3_resource.Bucket(bucket).download_file(obj_name, obj_name)
```

```python
download_impressions(bucket, year, month)
```
